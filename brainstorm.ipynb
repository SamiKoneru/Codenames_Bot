```python
import random
import sys
from pathlib import Path

import numpy as np

PROJECT_ROOT = Path("/Users/skon/Desktop/Projects/codenames")
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from game_logic.self_play_game_logic import SelfPlayCodenamesGame
from shared.oracle_clue import OracleClueConfig, generate_oracle_clue
from shared.training_shared import load_words

WORD_FILE = str(PROJECT_ROOT / "wordlist-eng.txt")
words = load_words(WORD_FILE)
game = SelfPlayCodenamesGame(words=words, seed=0)
state = game.reset_board()
print("Board sample:", state["board_words"][:5], "...")

# Synthetic board embeddings for oracle clue demo (replace with LLM/API vectors)
D = 32
rng = np.random.default_rng(0)
board_emb = rng.standard_normal((len(game.board_words), D)).astype(np.float32)
sp = game.get_spymaster_state()
oc = generate_oracle_clue(
    board_emb,
    sp["labels"],
    sp["revealed"],
    cfg=OracleClueConfig(),
    rng=random.Random(1),
)
print("Oracle clue k:", oc["k"], "clue_emb shape:", oc["clue_emb"].shape)
```
